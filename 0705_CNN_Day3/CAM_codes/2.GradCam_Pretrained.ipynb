# 2.GradCam_Pretrained.ipynb 說明
本 notebook 主要教學如何利用預訓練模型（如 ResNet、VGG 等）實作 Grad-CAM（Gradient-weighted Class Activation Mapping），用於視覺化 CNN 模型在影像分類時「關注」的區域。內容重點如下：
1. **預訓練模型載入**：使用 torchvision 等套件載入 ImageNet 預訓練的 CNN 模型。
2. **資料前處理**：對輸入影像進行 resize、normalize 等處理，符合模型需求。
3. **前向推論與特徵提取**：將影像輸入模型，取得分類結果與中間層特徵。
4. **Grad-CAM 計算與視覺化**：利用最後卷積層的梯度與特徵圖，計算出 Grad-CAM，並將其疊加在原圖上，視覺化模型「看」到的重點區域。
5. **應用與延伸**：可用於模型解釋、錯誤分析、醫學影像等領域。
Grad-CAM 是一種更通用的可解釋性技術，能用於多種 CNN 架構。本 notebook 適合用於理解深度學習模型的可解釋性（Explainable AI），並學習如何用 PyTorch 實作 Grad-CAM 技術。

# 二、Grad CAM

grad_cam.png

```python
class GradCam():
    def __init__(self, imgfile, model, img_size, label_dict, output_model_layer):
        self.imgfile = imgfile
        self.model = model
        self.model.eval()
        self.img_size = img_size
        self.label_dict = label_dict
        self.fmap_block = []
        self.grad_block = []
        
        # Regist hook for gradient
        output_model_layer.register_forward_hook(self.__forward_hook)   # feature map
        output_model_layer.register_backward_hook(self.__backward_hook) # gradient

        
    def ReadImage(self):
        img = cv2.imread(self.imgfile)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # BGR to RGB
        return img
    
    def processing_image(self):
        img = self.ReadImage()
        preprocess_fn = transforms.Compose([transforms.ToTensor(),
                                    transforms.Resize((self.img_size,self.img_size)),
                                    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),])
        ImgTest = preprocess_fn(img)
        
        # (3,224,224) -> (1,3,224,224)
        ImgTest = torch.unsqueeze(ImgTest, 0)
        # to gpu mode
        ImgTest = ImgTest.to(device)
        return ImgTest
    
    # 取得BP後的梯度
    def __backward_hook(self, module, grad_in, grad_out):
        self.grad_block.append(grad_out[0].detach())  # gradient
        
    # 取得feed forward的結果
    def __forward_hook(self, module, input, output):
        self.fmap_block.append(output)               # feature map
        
    def comp_class_vec(self, output_vec, index=None):
        """
        :param ouput_vec: tensor, softmax分數輸出
        :param index: int，類別, 預設為softmax結果最大輸出的
        :return: tensor, class vector
        """
        if not index:
            index = np.argmax(output_vec.cpu().data.numpy())
        else:
            index = np.array(index)
        index = index[np.newaxis, np.newaxis]
        index = torch.from_numpy(index)

        one_hot = torch.zeros(1, 1000).scatter_(1, index, 1) # 要調整，如果只有10個類別，請改成 torch.zeros(1, 10)
        one_hot = one_hot.to(device)
        one_hot.requires_grad = True
        class_vec = torch.sum(one_hot * output_vec)  # one_hot = 11.8605

        return class_vec

    def gradcam(self):
        x = self.processing_image()

        # 1. 模型預測，並儲存預測出來的結果
        Result = self.model(x)
        # 輸出softmax的結果，並將最大的output的argmax取出來
        arg_sm = np.argmax(Result.cpu().data.numpy())               
        pred_class_name = self.label_dict[arg_sm]
        
        # 2. Feed forward get feature map
        Feature_map = self.fmap_block[0].cpu().data.numpy().squeeze()
        print(Feature_map.shape)
        
        # 3. backpropogation, 計算模型預測出標籤的結果，並回推梯度
        self.model.zero_grad()
        class_loss = self.comp_class_vec(Result)
        class_loss.backward()
        # get back propogation gradient, (2048,7,7)
        grads_val = self.grad_block[0].cpu().data.numpy().squeeze()
        
        # 4. 將計算好的卷積gradient轉換成class weight
        # gradient mean
        weights = np.mean(grads_val, axis=(1, 2)) #(2048,7,7) -> (2048,)
        
        # 4. 將class weight的結果乘上Feature map上
        # 建立一個空的cam, 大小就是最後一層輸出的長與寬，如：(7,7) 
        cam = np.zeros(dtype = np.float32, shape = Feature_map.shape[1:3]) # cam shape (H, W)
        
        # use <pred_class> to choose which output label we want to visualize
        # 通常都以argmax(predsoutput)來參考
        for i, w in enumerate(weights):
            cam += w * Feature_map[i, :, :]
        
        ##先做一個relu再用minmax將數值normalize到0,1之間
        heatmap = np.maximum(cam, 0) / np.max(cam)
        
        return heatmap, pred_class_name
    

    def plot_heatmap(self):
        # 圖像讀取，並resize
        img = self.ReadImage()
        img = cv2.resize(img,(self.img_size, self.img_size))
        
        # 取得Cam的結果跟預測出來的class name
        heatmap, pred_class_name = self.gradcam()
        
        # 將heatmap 的數值從 (0,1) 變成 (0,255)
        heatmap = np.uint8(255 * heatmap)
        
        # heatmap resize
        jet_heatmap = cv2.resize(heatmap,(self.img_size, self.img_size))

        plt.figure(figsize=(15, 5))
        plt.subplot(1,3,1)
        plt.imshow(img)
        plt.title("Raw Image")

        plt.subplot(1,3,2)
        plt.imshow(img,alpha = 0.6)
        plt.imshow(jet_heatmap, cmap='jet', alpha=0.4)
        plt.title("Predict Result = %s"%(pred_class_name))

        plt.subplot(1,3,3)
        plt.imshow(jet_heatmap, cmap='jet')
        plt.title("Attention map")
        
        
```

```python
# 1. 輸入想要檢測的模型，這裡以訓練imagenet的resnet50模型做完範例
cnn_model = models.resnet50(pretrained=True)
cnn_model.to(device)

GRADCAM = GradCam(imgfile = Dogimgfile, 
                  model = cnn_model, 
                  img_size = img_size, 
                  label_dict = label_dict, 
                  output_model_layer = cnn_model.layer4[2].conv2)

GRADCAM.plot_heatmap()
```
